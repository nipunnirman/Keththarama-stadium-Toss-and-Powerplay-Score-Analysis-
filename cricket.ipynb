```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

# Load your CSV
df = pd.read_csv('keththarama.csv')

# Clean columns
df.columns = df.columns.str.strip()

# Infer bat/bowl first (as in analysis)
df['Bat_First_Team'] = df.apply(lambda row: row['Toss_Win'] if row['Decision'] == 'Bat' else row['Match'].replace(row['Toss_Win'] + ' vs ', '').replace(' vs ' + row['Toss_Win'], ''), axis=1)
df['Bowl_First_Team'] = df.apply(lambda row: row['Toss_Win'] if row['Decision'] == 'Bowl' else row['Match'].replace(row['Toss_Win'] + ' vs ', '').replace(' vs ' + row['Toss_Win'], ''), axis=1)

# Create target for toss decision: 1 if bat first wins
df['Bat_First_Win'] = (df['Winner'] == df['Bat_First_Team']).astype(int)
df = df[df['Winner'] != 'TIED']  # Exclude ties

# Encode categorical features (teams)
le = LabelEncoder()
df['Home_Team_Encoded'] = le.fit_transform(df['Match'].apply(lambda x: x.split(' vs ')[0]))  # Assuming first team is home (SL often)
df['Away_Team_Encoded'] = le.fit_transform(df['Match'].apply(lambda x: x.split(' vs ')[1]))

# Features for models
features = ['Home_Team_Encoded', 'Away_Team_Encoded']  # Add more like rankings later
```
